Delete Missing Tuples

```python
df = dataset.copy()

missing_duration = pd.isna(df["duration"])
missing_history = pd.isna(df["history"])
missing_purpose = pd.isna(df["purpose"])
missing_amount = pd.isna(df["amount"])
missing_savings = pd.isna(df["savings"])
missing_employ = pd.isna(df["employ"])
missing_status = pd.isna(df["status"])
missing_others = pd.isna(df["others"])
missing_property = pd.isna(df["property"])
missing_employ = pd.isna(df["employ"])
missing_age = pd.isna(df["age"])
missing_housing = pd.isna(df["housing"])
missing_cards = pd.isna(df["cards"])
missing_job = pd.isna(df["job"])
missing_liable = pd.isna(df["liable"])
missing_foreign = pd.isna(df["foreign"])

missingIndices = df[
    missing_duration |
    missing_history |
    missing_purpose |
    missing_amount |
    missing_savings |
    missing_employ |
    missing_status |
    missing_others |
    missing_property |
    missing_employ |
    missing_age |
    missing_housing |
    missing_cards |
    missing_job |
    missing_liable |
    missing_foreign
    ].index

df = df.drop(missingIndices, axis = 0)
print("Dimensions of ", dataset.shape," has been reduced to ", df.shape)
```

```python
df.head(1)
```

Handle Categorical Values

```python
df['checkingstatus1'] = df['checkingstatus1'].replace({
    'A11': 0, 
    'A12': 1,
    'A13': 2,
    'A14': 3
    })
df['history'] = df['history'].replace({
    'A30': 0, 
    'A31': 1, 
    'A32': 2, 
    'A33': 3, 
    'A34': 4
    })
df['purpose'] = df['purpose'].replace({
    'A40': 0, 
    'A41': 1, 
    'A42': 2, 
    'A43': 3, 
    'A44': 4, 
    'A45': 5, 
    'A46': 6, 
    'A47': 7,
    'A48': 8, 
    'A49': 9, 
    'A410': 10, 
    })
df['savings'] = df['savings'].replace({
    'A61': 0, 
    'A62': 1,
    'A63': 2, 
    'A64': 3, 
    'A65': 4
    })
df['employ'] = df['employ'].replace({
    'A71': 0, 
    'A72': 1,
    'A73': 2, 
    'A74': 3, 
    'A75': 4
    })
df['status'] = df['status'].replace({
    'A91': 0, 
    'A92': 1,
    'A93': 2, 
    'A94': 3, 
    'A95': 4
    })
df['others'] = df['others'].replace({
    'A101': 0, 
    'A102': 1,
    'A103': 2, 
    })
df['property'] = df['property'].replace({
    'A121': 0, 
    'A122': 1, 
    'A123': 2, 
    'A124': 3, 
    })
df['otherplans'] = df['otherplans'].replace({
    'A141': 0, 
    'A142': 1, 
    'A143': 2, 
    })
df['housing'] = df['housing'].replace({
    'A151': 0, 
    'A152': 1, 
    'A153': 2, 
    })
df['job'] = df['job'].replace({
    'A171': 0, 
    'A172': 1, 
    'A173': 2, 
    'A174': 3, 
    })
df['foreign'] = df['foreign'].replace({
    'A201': 0, 
    'A202': 1, 
    })
    
df.info()
```
